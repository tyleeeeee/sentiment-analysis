```python
import pandas as pd
import numpy as np
```

```python
traindata = pd.read_csv('train.csv', encoding= 'unicode_escape')
```

```python
testdata = pd.read_csv('test.csv', encoding= 'unicode_escape')
```

```python
traindata['category'].value_counts()
```

```python
import nltk
from nltk.corpus import stopwords
```

```python
stop = set(stopwords.words('english'))
#remove no, not in stop
stop.remove('no') 
stop.remove('not')
#stop
```

```python
#Lemmatization 
from nltk.stem import PorterStemmer
def stem(str_):
    ps = PorterStemmer()
    return ps.stem(str_)

#Removing nonword
import string
import re
def remove_nonword(str_):
    return re.sub(r'[^a-z]', ' ', str_) 

#lower
def make_lower(str_):
    return str_.lower()
```

traindata preprocessing

```python
from nltk import word_tokenize
data = []
for item in traindata['content']:
    word = ''
    item = stem(item)
    item = remove_nonword(item)
    item = make_lower(item)
    item = word_tokenize(item)
    #print(item)
    for i in range(len(item)):
        if item[i] not in stop:
            if len(item[i]) > 3:
                word = word + item[i] + ' '
    data.append(word)
#data
```

testdata preprocessing

Vectorizer

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer(max_features=2500)
X = vectorizer.fit_transform(data)
```

split data to train data and test data

```python
from sklearn.model_selection import train_test_split
y = traindata['category']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score

lr = LogisticRegression(C = 0.01)
lr.fit(X_train, y_train)
accuracy = accuracy_score(y_test, lr.predict(X_test))
f1 = f1_score(y_test, lr.predict(X_test), average='macro')

print(accuracy)
print(f1)
```

```python
tdata = []
for item in testdata['content']:
    word = ''
    item = stem(item)
    item = remove_nonword(item)
    item = make_lower(item)
    item = word_tokenize(item)
    #print(item)
    for i in range(len(item)):
        if item[i] not in stop:
            if len(item[i]) > 3:
                word = word + item[i] + ' '
    tdata.append(word)
#data
```
